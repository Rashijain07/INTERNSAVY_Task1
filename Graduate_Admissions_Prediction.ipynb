Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load dataset

```python
dataset=pd.read_csv("Admission_Predict.csv")
dataset
```

Summarize dataset

```python
print(dataset.head())
print(dataset.shape)
```

Segregate dataset into X & Y

```python
X=dataset.iloc[:,:-1].values
X
```

```python
Y=dataset.iloc[:,-1].values
Y
```

  Converting Y values into 0 and 1

```python
import numpy as np

def convert_to_binary(data, threshold):
    binary_data = np.where(data < threshold, 0, 1)
    return binary_data


values = np.array(Y)
threshold = 0.5

binary_values = convert_to_binary(values, threshold)
print(binary_values)
```

```python
Y=binary_values.reshape(len(Y),1)
Y
```

Splitting dataset into train & test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.25, random_state=0)
print(X_train)
print(X_test)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
X_test
```

Training

```python
import warnings
warnings.simplefilter("ignore")

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train, y_train)
```

Prediction

```python
input=[[1, 223, 175, 3, 13, 45, 10, 1]]
result= model.predict(sc.transform(input))
print(result)
if(result==1):
  print("Chance of taking admission is high")
else:
  print("Chance of taking admission is low")
```

Evaluation

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score is : {0}% ".format(accuracy_score(y_pred, y_test)*100))
```
